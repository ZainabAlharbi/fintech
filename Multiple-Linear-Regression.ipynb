```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn import linear_model
```

```python
np.random.seed(0) # same to 
x1 = 3.0 * np.random.randn(100) + 2.0
x2 = 3.5 * np.random.randn(100) + 1.0
res = 0.5 * np.random.randn(100) # e
y = 1.5 + 0.45 * x1 + res


df = pd.DataFrame({'x1': x1, 'x2': x2, 'y': y})
df.head()
```

```python
X = df[['x1', 'x2']]
print(df[['x1','x2']])
```

```python
y = df['y']
print(y)
```

```python
regression = linear_model.LinearRegression()
```

```python
regression.fit(X, y)
```

```python
print(regression.coef_)
print(regression.predict([[0, 0]]))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['x1'], df['x2'], df['y'], c = 'r', marker = 'o')
xx1, xx2 = np.meshgrid(range(-5, 8), range(-6, 9))
yy = -45.25*xx1 - 0.886*xx2 + 169.23
ax.plot_surface(xx1, xx2, yy, alpha=0.5)
plt.show()
```
